```python
def stock_return(price_now, price_past):
    try:
        ret =  (price_now-price_past) / price_past
    except:
        ret = np.nan
    return ret
```

```python
def get_weekly_return(a_df):
    date_list = a_df.iloc[:,0:1].values.tolist()
    price_list =a_df.iloc[:,1:2].values.tolist()
    
    return_list = []
    today = -1
    yesterday = -1
    this_week_last_price = 0
    last_week_last_price = price_list[0]
    
    tem_price_holer = []# nan
    
    for date, price in zip(date_list, price_list):
        today = date[0].weekday()
        
        if today < yesterday: #monday 0, sunday 6 todayweekday
            
            if len(tem_price_holer) != 0:
                this_week_last_price = tem_price_holer[-1]
                week_ret = stock_return(this_week_last_price, last_week_last_price)
                return_list.append(week_ret)
                
                last_week_last_price = this_week_last_price
            else:
                return_list.append(np.nan)
                last_week_last_price = np.nan
                
            tem_price_holer = [] #
            
        if not math.isnan(price[0]): #price nan
            tem_price_holer.append(price[0])
        
        yesterday = today
        

    if len(tem_price_holer) != 0:
        this_week_last_price = tem_price_holer[-1]
        week_ret = stock_return(this_week_last_price, last_week_last_price)
        return_list.append(week_ret)
    else:
        return_list.append(np.nan)
        last_week_last_price = np.nan
    
    return return_list
```

# performance measures

## ASKSR

```python
from scipy.stats import  skew, kurtosis
import scipy
def get_asksr(ret_list, rf = 0.0104):
    #paremeters of stats
    mean = sum(ret_list)  /  len(ret_list) 
    std = np.std(ret_list)
    S = skew(ret_list)
    K = kurtosis(ret_list, fisher = False )
    print('Condition:',  K > 3 + 5/3 * S * S)
    if K < 3 + 5/3 * S * S:
        print(S, K)
        return np.nan
    
    # other parameters facilitating caculation
    alpha = 3*math.sqrt(3*K-4*(S**2) -9) / (std**2) / (3*K-5*(S**2) -9) 
    beta = 3 * S / (std) / (3*K-5*(S**2) -9)  
    eta = mean - 3 * S * (std) / (3*K-4*(S**2) -9)  
    delta = 3 * std * math.sqrt(3*K-5*(S**2) -9) / (3*K-4*(S**2) -9) 
    
    
    a_lambda = beta + alpha*(eta - rf) / math.sqrt(delta*delta + (eta-rf) * (eta-rf) )
    
    #
    ASKSR_square = 2* ( a_lambda*(eta-rf) - delta*(math.sqrt(alpha**2 - beta**2 ) - math.sqrt(alpha**2 - (beta -a_lambda )**2)  )          )
    return math.sqrt( ASKSR_square )
```

## omega

```python
def get_omega(ret_list, L = 0.0104, ):

    lose_list = []
    for ret in ret_list:

        if ret < L:
            lose_list.append( L - ret  )
    mean = sum(ret_list) / len(ret_list)
    expected_lose =  ( sum(lose_list) / len(lose_list) )
    return ( mean -  L) / expected_lose
```

## riskiness R

```python
import math
from scipy.optimize import fsolve
```

```python
def get_riskiness_r(ret_list):
    # intitail alpha  = mean
    alpha =  100 * sum(ret_list)  /  len(ret_list) 
    n  = len(ret_list)
    
    def f(alpha):
        exp_list1 = list( map(lambda x : math.exp(-1* alpha* x), ret_list) ) 
        
        return   math.pow(n - sum(exp_list1) , 2)  



    

    alpha_star = fsolve(f, alpha)

        
    return math.exp( -1 * alpha_star   )
```

# sort etf ranking by those measures

```python
def get_ranking(data_type = 'week'):
    data_type = data_type.lower()
    
    data_asksr = {}
    data_omega = {}
    data_riskiness = {}
    if data_type == 'week':
        ret_df = df_week_ret
    elif data_type == 'month':
        ret_df = df_month_ret
    else:
        return 'Wrong input.'
    
    for i in range(1,33):
        a_id = id_list[i]
        ret_list = list(ret_df[a_id].dropna())
        
        measure_asksr = get_asksr(ret_list)
        measure_omega = get_omega(ret_list, L = 0.0104)
        measure_riskiness = get_riskiness_r(ret_list)
        


        #replace ret dataframe as weekly return

        data_asksr[a_id] = [measure_asksr]
        data_omega[a_id] = [measure_omega]
        data_riskiness[a_id] = [measure_riskiness]
        
    data_list = [data_asksr, data_omega, data_riskiness]
    columns_name =  ['asksr', 'omega', 'riskiness']
    output_data = {}
    # output: dataframe of ranking
    for i in range(3):
        data = data_list[i]
        
        ranking_df = pd.DataFrame(data = data).T
        ranking_df.columns = [data_type]
        if i !=2:
            ranking_df = ranking_df.sort_values(by=[ data_type ],  ascending=False)
        else:
            ranking_df = ranking_df.sort_values(by=[ data_type ],  ascending=True)                                                    
        
        # save the ranking as a dictionary
        output_data[ columns_name[i]] = list(ranking_df.index)
        
    #then transform the dictionary to a dataframe
    return     pd.DataFrame(data = output_data)
```

## print spearman rank correlation

```python
from scipy.stats.stats import spearmanr
def rank_corr(ranklist_a, ranklist_b):
    rank_a = []
    rank_b = []
    for id in id_list[1:]:
        rank_a.append(ranklist_a.index(id) )
        rank_b.append(ranklist_b.index(id) )
    
    
    
    corr, _= spearmanr(rank_a, rank_b)
    return corr
```

```python
def print_correlation_metric(a_type = 'week'):
    
    def print_cross():
        data = {}
        
        data['asksr'] =  rank_corr( week_ranking['asksr'].tolist(), month_ranking['asksr'].tolist())
        data['omega'] =  rank_corr( week_ranking['omega'].tolist(), month_ranking['omega'].tolist())
        data['riskiness'] =  rank_corr( week_ranking['riskiness'].tolist(), month_ranking['riskiness'].tolist())
                
        print( pd.DataFrame(data = data, index = ['cross']) )
    
    if a_type == 'week':
        a_df = week_ranking
    elif a_type == 'month':
        a_df = month_ranking
    elif a_type == 'cross': #cross
        return print_cross()
    
    data = {}
    
    for i in ['asksr', 'omega', 'riskiness']:
        data[i] = []
        for j in ['asksr', 'omega', 'riskiness']:
            data[i].append(     rank_corr( a_df[i].tolist(), a_df[j].tolist())                 )
   
    df = pd.DataFrame(data = data)
    df.index = ['asksr', 'omega', 'riskiness']
    print(df)
```
