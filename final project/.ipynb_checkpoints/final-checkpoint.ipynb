# 基金相關係數資料庫


### 基本介紹
投資人輸入自己已擁有的基金投資組合，程式推薦他和此投資組合相關係數最小的基金。

### 預期目標
希望幫助投資人在投資基金時能夠更加地分散風險。

```python
import math
def stock_return(price_now, price_past):
    try:
        ret =  (price_now-price_past) / price_past
    except:
        ret = np.nan
    return ret
```

## weekly return

```python
def get_weekly_return(a_df):
    date_list = a_df.iloc[:,0:1].values.tolist()
    price_list =a_df.iloc[:,1:2].values.tolist()
    index_list = []
    
    return_list = []
    today = -1
    yesterday = -1
    this_week_last_price = 0
    last_week_last_price = price_list[0]
    
    tem_price_holer = []# 紀錄這周所有價格，為了判斷是否都是nan
    
    for date, price in zip(date_list, price_list):
        today = date[0].weekday()
        
        if today < yesterday: #由於monday 是0, sumday 是6 如果today的weekday比較小，代表換周了
            
            if len(tem_price_holer) != 0:
                this_week_last_price = tem_price_holer[-1]
                week_ret = stock_return(this_week_last_price, last_week_last_price)
                #append
                return_list.append(week_ret)
                index_list.append(date[0])
                #update
                last_week_last_price = this_week_last_price
            else:
                return_list.append(np.nan)
                index_list.append(date[0])
                last_week_last_price = np.nan
                
            tem_price_holer = [] #更新
            
        if not math.isnan(price[0]): #price 不是nan
            tem_price_holer.append(price[0])
        
        yesterday = today
        
    if len(tem_price_holer) >0:
        this_week_last_price = tem_price_holer[-1]
    else:
        this_week_last_price= np.nan
    
    return return_list, index_list, last_week_last_price, this_week_last_price, date[0]
```

## monthly return

```python
def get_monthly_return(a_df):
    date_list = a_df.iloc[:,0:1].values.tolist()
    price_list =a_df.iloc[:,1:2].values.tolist()
    index_list = []
    return_list = []
    today = -1
    yesterday = -1
    this_month_last_price = 0
    last_month_last_price = price_list[0]
    
    tem_price_holer = []# 紀錄這月所有價格，為了判斷是否都是nan
    
    for date, price in zip(date_list, price_list):
        today = date[0].day
        
        if today < yesterday: #由於1號 是1, 31號是31 如果today比較小，代表換月了
            
            if len(tem_price_holer) != 0:
                this_month_last_price = tem_price_holer[-1]
                month_ret = stock_return(this_month_last_price, last_month_last_price)
                return_list.append(month_ret)
                index_list.append(date[0])
                
                last_month_last_price = this_month_last_price
            else:
                return_list.append(np.nan)
                index_list.append(date[0])
                last_month_last_price = np.nan
                
            tem_price_holer = [] #更新
            
        if not math.isnan(price[0]): #price 不是nan
            tem_price_holer.append(price[0])
        
        yesterday = today

    if len(tem_price_holer) >0:
        this_month_last_price = tem_price_holer[-1]
    else:
        this_month_last_price= np.nan
        
    return return_list, index_list, last_month_last_price, this_month_last_price , date[0]
```

```python
data2 = {}
for i in range(1, 6193):
    ret_list, index_list, last_month_last_price, this_month_last_price, _  = get_monthly_return(ret_df.iloc[:,[0,i]])
    
    #replace ret dataframe as weekly return
    a_id = id_list[i]
    data2[a_id] = ret_list
    data_price[a_id] += [last_month_last_price, this_month_last_price]
```

```python
from correlation import *

def get_corr(a, b, period=100):
    corr={}
    corr['person_corr']=[person_corr(df_week_ret, a, b, period), person_corr(df_month_ret, a, b, period)]
    corr['rank_corr']=[rank_corr(df_week_ret, a, b, period), rank_corr(df_month_ret, a, b, period)]
    corr['downside_corr']=[downside_corr(df_week_ret, a, b, period), downside_corr(df_month_ret, a, b, period)]
    return pd.DataFrame(data = corr, index = ['week','month'])

```

```python
def get_top(func, fundlist, fund, top):
    fund_corr = {}
    for fundname in ret_df.columns[1:]:
        if fund == fundname:
            continue
        corr = func(fundlist, fund, fundname)
        if not math.isnan(corr):
            fund_corr[fundname] = corr
    fund_corr_list = [ it[0] for it in sorted(fund_corr.items(), key=lambda d: d[1])]
    top_list = ['nan' for i in range(top)]
    for i in range(min(top, len(fund_corr_list))):
        top_list[i] = fund_corr_list[i]
    return top_list

def recommand_funds(fund, top=10):
    ranklist = {}
    ranklist['week_person'] = get_top(person_corr, df_week_ret, fund, top)
    ranklist['month_person'] = get_top(person_corr, df_month_ret, fund, top)
    ranklist['week_rank'] = get_top(rank_corr, df_week_ret, fund, top)
    ranklist['month_rank'] = get_top(rank_corr, df_month_ret, fund, top)
    ranklist['week_downside'] = get_top(downside_corr, df_week_ret, fund, top)
    ranklist['month_downside'] = get_top(downside_corr, df_month_ret, fund, top)
    return pd.DataFrame(data=ranklist)
```

# 使用資料庫

```python
import sqlite3
import pandas as pd
from scipy.stats.stats import pearsonr
from datetime import datetime
import time
import numpy as np
```

```python
def downside_corr(a,b):
    i = a[1:]<a[:-1]
    corrab, _ = pearsonr(a[1:][i], b[1:][i])
    j = b[1:]<b[:-1]
    corrba, _ = pearsonr(a[1:][j], b[1:][j])
    return (corrab+corrba)/2
```

```python
cor_list = []
id_list = []
mean_list = []
db = sqlite3.connect('data/fund.sqlite')
cur = db.cursor()
cur.execute('select * from week')
r = cur.fetchall()
for i in range(len(r)):
    id_list.append(r[i][0])

# get rf
df_rf  = pd.read_csv('data/rf.csv')
time_list = df_rf['y/m'].tolist()
rf_list = [float(x)/5200 for x in df_rf['rf'].tolist()]

time_to_rf = dict()
for i in range(len(time_list)):
    year, month = time_list[i].split('/')
    year = int(year) + 1911
    month = int(month)
    string = str(year) +'-%02d'%month #%Y-%m
    time_to_rf[string] = rf_list[i]
    
cur.execute('PRAGMA  table_info([week])')
rf_array = []
r = cur.fetchall()
for i in range(1, len(r)):
    date = r[i][1][:-3]
    rf_array.append( time_to_rf[date]  )
rf_array = np.array(rf_array)
```

```python
def get_corr(fundA, fundB, period=100):
    cur.execute('select * from week WHERE "index" = "'+fundA+'"')
    r = cur.fetchall()
    x = np.array(r[0][1:], dtype= 'float64').reshape(1,-1)
    
    cur.execute('select * from week WHERE "index" = "'+fundB+'"')
    r = cur.fetchall()
    y = np.array(r[0][1:], dtype= 'float64').reshape(1,-1)
    
    z = np.concatenate((x,y), axis=0).T
    rf_new = rf_array[~np.isnan(z).any(axis=1)]
    z_new = z[~np.isnan(z).any(axis=1)].T
    x_new = z_new[0] - rf_new
    y_new = z_new[1] - rf_new
    cor1,_ = pearsonr(x_new, y_new)
    cor2 = downside_corr(x_new, y_new)
    print('一般相關係數:',cor1,'下跌相關係數',cor2)
```

```python
while(1):
    funds = input('請輸入要比較的基金(fundA fundB):').split()
    if funds == ['quit']:
        break
    if len(funds) != 2:
        print('請輸入兩個基金並以單一空格隔開')
        continue
    else:
        fundA, fundB = funds
        correct_name = True
        if fundA not in id_list:
            correct_name = False
            print(fundA,'找不到!')
        if fundB not in id_list:
            correct_name = False
            print(fundB,'找不到!')
        if correct_name:
            get_corr(fundA, fundB)
        else:
            continue
```

```python
def recommend_funds(fund):
    fund_list = []
    mean_list = []
    cor1_list = []
    cor2_list = []
    
    cur.execute('select * from week')
    r = cur.fetchall()
    p = id_list.index(fund)
    x = np.array(r[p][1:], dtype= 'float64').reshape(1,-1)
    for i in range(len(r)):
        if i==p:
            continue
        fund_list.append(r[i][0])
        y = np.array(r[i][1:], dtype= 'float64').reshape(1,-1)
        z = np.concatenate((x,y), axis=0).T
        rf_new = rf_array[~np.isnan(z).any(axis=1)]
        z_new = z[~np.isnan(z).any(axis=1)].T
        x_new = z_new[0] - rf_new
        y_new = z_new[1] - rf_new
        y = y.reshape(-1,)
        y = y[~np.isnan(y)]
        mean_list.append(y.mean()*4)
        cor1,_ = pearsonr(x_new, y_new)
        cor1_list.append(cor1)
        cor2 = downside_corr(x_new, y_new)
        cor2_list.append(cor2)

    print('建議您在以下基金中選擇一項購買:')
    print('考慮一般相關性:')
    cor1_list = np.array(cor1_list)
    indices = cor1_list.argsort()[:6]
    output = []
    for i in indices[0:3]: 
        output.append( (mean_list[i], 'A+', id_list[i]) )
    for i in indices[3:]: 
        output.append( (mean_list[i], 'A', id_list[i]) )

    output.sort(reverse = True)
    print('{0:<3s} {1:<12s} {2:<5s} {3:<6s}'.format('排名','基金ID','平均獲利','分散風險程度'))
    for rank, (mean, degree, fund_id) in enumerate(output):
        print('{0:<5d} {1:<14s} {2:<.6f} {3:<6s}'.format(rank+1,fund_id,mean,degree))
        
    print('考慮下跌相關性:')
    cor2_list = np.array(cor2_list)
    indices = cor2_list.argsort()[:6]
    output = []
    for i in indices[0:3]: 
        output.append( (mean_list[i], 'A+', id_list[i]) )
    for i in indices[3:]: 
        output.append( (mean_list[i], 'A', id_list[i]) )

    output.sort(reverse = True)
    print('{0:<3s} {1:<12s} {2:<5s} {3:<6s}'.format('排名','基金ID','平均獲利','分散風險程度'))
    for rank, (mean, degree, fund_id) in enumerate(output):
        print('{0:<5d} {1:<14s} {2:<.6f} {3:<6s}'.format(rank+1,fund_id,mean,degree))
```

```python
while(1):
    fund = input('請輸入目前的基金:')
    if fund == 'quit':
        break
    if fund not in id_list:
        print(fund,'找不到!')
        continue
    else:
        recommend_funds(fund)
```

```python
eps = 1e-7
picked_fund = []
picked_num = []

print('請輸入基金ID與投資比例(或金額)')
finish = False
while not finish:
    fund_name = input('請輸入基金ID(或finish結束輸入):')
    if fund_name in id_list:
        while True:
            fund_num = input('請輸入投資比例(或金額):')
            try:
                fund_num = float(fund_num)
                if fund_num > -eps:
                    picked_fund.append(fund_name)
                    picked_num.append(fund_num)
                    break
                else:
                    print('請輸入>=0的數')
            except:
                print('請輸入>=0的數')
    elif fund_name=='finish':
        if len(picked_fund)==0:
            print('您並未輸入投資組合')
        elif sum(picked_num) < eps:
            print('您輸入的數字過低')
        else:
            finish = True
            break
    else:
        print('找不到',fund_name)
            
fund_list = []
mean_list = []
cor1_list = []
cor2_list = []
picked_id = []
for fund_name in picked_fund:
    picked_id.append(id_list.index(fund_name))

cur.execute('select * from week')
r = cur.fetchall()
picked_array = []
for p in picked_id:
    picked_array.append(np.array(r[p][1:], dtype='float64'))
x = np.array(picked_array)
total_weight = sum(picked_num)
weight = np.array(picked_num).reshape(-1,1)
for i in range(len(r)):
    if i in picked_id:
        continue
    fund_list.append(r[i][0])
    y = np.array(r[i][1:], dtype= 'float64').reshape(1,-1)
    z = np.concatenate((x,y), axis=0).T
    rf_new = rf_array[~np.isnan(z).any(axis=1)]
    z_new = z[~np.isnan(z).any(axis=1)].T
    x_new = z_new[:-1]*weight
    x_new = x_new.sum(axis=0)/total_weight - rf_new
    y_new = z_new[-1] - rf_new
    y = y.reshape(-1,)
    y = y[~np.isnan(y)]
    mean_list.append(y.mean()*4)
    cor1,_ = pearsonr(x_new, y_new)
    cor1_list.append(cor1)
    cor2 = downside_corr(x_new, y_new)
    cor2_list.append(cor2)
    
print('建議您在以下基金中選擇一項購買:')
print('考慮一般相關性:')
cor1_list = np.array(cor1_list)
indices = cor1_list.argsort()[:6]
output = []
for i in indices[0:3]: 
    output.append( (mean_list[i], 'A+', id_list[i]) )
for i in indices[3:]: 
    output.append( (mean_list[i], 'A', id_list[i]) )

output.sort(reverse = True)
print('{0:<3s} {1:<12s} {2:<5s} {3:<6s}'.format('排名','基金ID','平均獲利','分散風險程度'))
for rank, (mean, degree, fund_id) in enumerate(output):
    print('{0:<5d} {1:<14s} {2:<.6f} {3:<6s}'.format(rank+1,fund_id,mean,degree))

print('考慮下跌相關性:')
cor2_list = np.array(cor2_list)
indices = cor2_list.argsort()[:6]
output = []
for i in indices[0:3]: 
    output.append( (mean_list[i], 'A+', id_list[i]) )
for i in indices[3:]: 
    output.append( (mean_list[i], 'A', id_list[i]) )

output.sort(reverse = True)
print('{0:<3s} {1:<12s} {2:<5s} {3:<6s}'.format('排名','基金ID','平均獲利','分散風險程度'))
for rank, (mean, degree, fund_id) in enumerate(output):
    print('{0:<5d} {1:<14s} {2:<.6f} {3:<6s}'.format(rank+1,fund_id,mean,degree))
```
